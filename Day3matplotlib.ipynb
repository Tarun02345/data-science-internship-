```python

import numpy as np
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [23, 17, 35, 29]

# Create an array for the positions on the x-axis
x_pos = np.arange(len(categories))

# Create the bar graph
plt.bar(x_pos, values, align='center', alpha=0.7)

# Add titles and labels
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph Example')

# Add the category names as x-axis labels
plt.xticks(x_pos, categories)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
#frequencies
ages=[2,5,70,40,30,45,50,45,43,40,44,60,7,13,57,18,90,77,32,21,40]
#setting the ranges and no.of intervals
range=[0,100]
bins =10
#plotting a histogram
plt.hist (ages, bins, range,
color="green", histtype="bar",edgecolor="black")
plt.xlabel('age')
#frequency label
plt.ylabel('No. of people')
# plot title
plt.title('My histogram')
```

```python
import matplotlib.pyplot as plt
#frequencies
ages=[2,5,70,40,30,45,50,45,43,40,44,60,7,13,57,18,90,77,32,21,40]
# setting the ranges and no. of intervals
range = (0, 100)
bins =10
#plotting a histogram
plt.hist (ages, bins, range,
color="red", histtype="bar",edgecolor="gold",alpha=1)
plt.xlabel('age')
#frequency label
plt.ylabel('No. of people')
# plot title
plt.title('My histogram')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('Data.csv')

# Extract data for plotting
x = df['Age']
y = df['Salary']

# Create the bar graph
plt.bar(x, y)

# Add titles and labels (optional)
plt.xlabel('Age')
plt.ylabel('Salry')
plt.title('Salary with age')

# Display the plot
plt.show()
```
